## Generating Crystal System Labels from Raw .cif Files

```julia
rm("../labels/graph_reader_test.txt")
io = open("../labels/graph_reader_test.txt", "w")

# Write .cif file header
io.write("cif, crystalsystem")
```

```julia
for (root, dirs, files) in walkdir("/home/qingyanz/cod/cif/")
    # cifs appear only in directories that are numbers
    filter!(dir->(tryparse(Int32, dir) == nothing)!=true, dirs)
    
    # If down to the lowest level
    if isempty(dirs)
        for file in files
            # openfile = open(string(root, "/", file))
            # iterfile = eachline(openfile)
            
            for line in collect(eachline(string(root, "/", file)))
                if startswith(line, "_symmetry_cell_setting")
                    cifnum = chop(file, tail=4)
                    println(string(cifnum, ", ", split(line)[2]))
                    write(io, string("\n", cifnum, ", ", split(line)[2]))
                    # close(openfile)
                    break
                    
                elseif startswith(line, "_space_group_crystal_system")
                    cifnum = chop(file, tail=4)
                    println(string(cifnum, ", ", split(line)[2]))
                    write(io, string("\n", cifnum, ", ", split(line)[2]))
                    # close(openfile)
                    break
                end
                ## elseif: infer crystal system label from index number?
            end            
        end
    end
end
```
